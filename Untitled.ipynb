```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline


df = pd.read_csv(r"cleaned_dataset.csv")

# Split data
X_train, X_test, y_train, y_test = train_test_split(df["text"], df["account_type"], test_size=0.2, random_state=42)

# Create a text classification model
model = make_pipeline(TfidfVectorizer(), MultinomialNB())

# Train the model
model.fit(X_train, y_train)

# Test with new text
new_text = ["This is how I usually write."]
prediction = model.predict(new_text)

print("Written by you?" if prediction[0] == 1 else "Not written by you")

```
