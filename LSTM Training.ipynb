```python
import pandas as pd
import tensorflow
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.regularizers import l2
from tensorflow.keras.layers import Bidirectional
```

```python
data=pd.read_csv(r"cleaned_dataset.csv")
```

```python
data
```

```python
dept=["bot","human"]
```

```python
data["account_type_id"]=data["account_type"].apply(lambda x: dept.index(x))
```

```python
data["account_type_id"].value_counts()
```

```python
data['text'] = data['text'].astype(str)
```

```python
data["text"]=data["text"].apply(lambda x: x.replace("(","").replace(")","").replace("\'","").replace("-","").replace(",","").replace(":","").replace(".","").replace("!",""))
```

```python
data
```

```python
tokenizer = Tokenizer(num_words=5000)
```

```python
tokenizer.fit_on_texts(data["text"])
```

```python
tokenizer.word_index
```

```python
sequence = tokenizer.texts_to_sequences(data["text"])
```

```python
sequence[0]
```

```python
maxlen=0
for i in sequence:
    if len(i)>maxlen:
        maxlen=len(i)
maxlen
```

```python
padded=pad_sequences(sequence,maxlen=46,padding="post")
```

```python
len(padded[0])
```

```python
data
```

```python
y=data.iloc[:,-1]
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(padded,y, test_size=0.2, random_state=42)
```

```python
len(X_train), len(y_train), len(X_test), len(y_test)
```

```python
model = Sequential()
model.add(Embedding(input_dim=5000, output_dim=64, input_length=46))  # Adjusted input_length
model.add(LSTM(units=128, dropout=0.3, recurrent_dropout=0.3))
model.add(Dense(64, activation='relu', kernel_regularizer=l2(0.01)))
model.add(Dense(1, activation='sigmoid'))  # Adjusted for 12 classes

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
batch_size=64
model.build(input_shape=(batch_size, 46))
model.summary()
```

```python
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True) 
```

```python
model.fit(X_train,y_train,epochs=25,batch_size=64,validation_split=0.2,callbacks=early_stopping)
```

```python
model.evaluate(X_test, y_test)
```
